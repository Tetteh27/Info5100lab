```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Load dataset
df = pd.read_csv("winequality-red.csv")  # Make sure this file is in your working directory

# Create output folder if it doesn't exist
os.makedirs("charts", exist_ok=True)

# Line plot: average alcohol by wine quality
line_data = df.groupby("quality")["alcohol"].mean()
plt.figure(figsize=(8,5))
line_data.plot(marker='o')
plt.title("Average Alcohol Content by Wine Quality")
plt.xlabel("Wine Quality")
plt.ylabel("Average Alcohol")
plt.grid(True)

# Save and show plot
plt.savefig("charts/line_plot_alcohol_vs_quality.png")
plt.show()
```

```python
import matplotlib.pyplot as plt
import os

# Create output folder if it doesn't exist
os.makedirs("charts", exist_ok=True)

# Histogram: distribution of citric acid
plt.figure(figsize=(8,5))
plt.hist(df["citric acid"], bins=20, color='skyblue', edgecolor='black')
plt.title("Histogram of Citric Acid")
plt.xlabel("Citric Acid")
plt.ylabel("Frequency")

# Save and show plot
plt.savefig("charts/histogram_citric_acid.png")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import os

# Create output folder if it doesn't exist
os.makedirs("charts", exist_ok=True)

# Box plot: alcohol content by wine quality
plt.figure(figsize=(8,5))
sns.boxplot(x="quality", y="alcohol", data=df)
plt.title("Box Plot of Alcohol Content by Wine Quality")
plt.xlabel("Wine Quality")
plt.ylabel("Alcohol Content")

# Save and show plot
plt.savefig("charts/boxplot_alcohol_vs_quality.png")
plt.show()
```

```python
import seaborn as sns
import os

# Create output folder if it doesn't exist
os.makedirs("charts", exist_ok=True)

# Pair plot: relationships among selected features
pair_df = df[["alcohol", "density", "pH", "quality"]]
sns.pairplot(pair_df, hue="quality", palette="husl")

# Save and show plot
plt.savefig("charts/pairplot_selected_features.png")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import os

# Create output folder if it doesn't exist
os.makedirs("charts", exist_ok=True)

# Compute correlation matrix
corr_matrix = df.corr(numeric_only=True)

# Set up the matplotlib figure
plt.figure(figsize=(12,8))

# Draw the heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Wine Features")

# Save and show plot
plt.savefig("charts/multivariate_correlation_heatmap.png")
plt.show()
```
